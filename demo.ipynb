```python
%pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(10)
y = x**2
print(x, y)
plt.scatter(x, y)
plt.show()
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y,'r-*') # red color, circle marker
plt.show()
```

```python
plt.plot(x, y, 'r-*')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X vs Y')
plt.show()
```
