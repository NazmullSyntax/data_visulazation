```python
import matplotlib.pyplot as plt


# Data
x = [1,2,3,4,5]

y1 = [10,20,30,40,50]

y2 = [50,40,30,20,10]


# create subplot
plt.subplot(1,2,1)   # 1 row, 2 columns, first plot
plt.plot(x,y1)

plt.title("Increasing")


plt.subplot(1,2,2)   # second plot
plt.plot(x,y2)

plt.title("Decreasing")


plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y1, 'o-')
plt.subplot(2, 1, 2)
plt.plot(x, y2, 'r-')
```
