```python
import matplotlib.pyplot as plt


# Data
x = [1,2,3,4,5]

y1 = [10,20,30,40,50]

y2 = [50,40,30,20,10]


# create subplot
plt.subplot(1,2,1)   # 1 row, 2 columns, first plot
plt.plot(x,y1)

plt.title("Increasing")


plt.subplot(1,2,2)   # second plot
plt.plot(x,y2)

plt.title("Decreasing")


plt.show()
```

```python
plt.subplot(4, 5, 1)
plt.plot(x, y1, 'o-')
plt.subplot(3, 5, 2)
plt.plot(x, y2, 'r-')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # [left, bottom, width, height]
ax1.plot(x, 'o-')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_title('Main Plot')
plt.show()
fig.savefig('my_figure.pdf')  # Save the figure as a PDF file
```

```python
fig,axes = plt.subplots(2, 2, figsize=(10, 10))
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
y1 = x * 10
y2 = y1[::-1]

axes[0, 0].plot(x, y1, 'o-')
axes[0, 0].set_title('Plot 1')
axes[0, 1].plot(x, y2, 'r-')
axes[0, 1].set_title('Plot 2')
axes[1, 0].plot(x, y1, 'g-')
axes[1, 0].set_title('Plot 3')
axes[1, 1].plot(x, y2, 'm-')
axes[1, 1].set_title('Plot 4')
plt.tight_layout()
plt.show()
axes[0, 0].set_title('Plot 1')
axes[0, 1].plot(x, y2, 'r-')
axes[0, 1].set_title('Plot 2')
axes[1, 0].plot(x, y1, 'g-')
axes[1, 0].set_title('Plot 3')
axes[1, 1].plot(x, y2, 'm-')
axes[1, 1].set_title('Plot 4')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(x, x + 1, color='r', linewidth=0.20)
axes[0, 1].plot(x, x + 2, color='g', linewidth=0.50)
axes[1, 0].plot(x, x + 3, color='b', linewidth=0.75)
axes[1, 1].plot(x, x + 4, color='m', linewidth=0.90)
plt.show()
```
